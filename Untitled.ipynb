```python
data=pd.read_csv('soc-sign-bitcoinalpha.csv') 
data=data.loc[data.SOURCE<500,:]
data=data.loc[data.TARGET<500,:]
data.reset_index(drop=True, inplace=True)
print(data)
#print(data.nrows)
#print(data[0:3]['TARGET'])
data.columns

```

```python
##Degree centrality

def rellenarGrafo(G,data):
    tam=data.shape[0]
    print(tam)
    for i in range(tam-1):
        #print(data.loc[i,'TARGET'])
        G.add_nodes_from([data.loc[i,'SOURCE'],data.loc[i,'TARGET']])
        G.add_edges_from([(data.loc[i,'SOURCE'], data.loc[i,'TARGET'])])
        
def addDegreeCentrality(G,data):
    dc = nx.degree_centrality(G)
    #print(dc)
    for it in dc:
        #print(it)
        #print(dc[it])
        data.loc[data.SOURCE==it, 'SOURCE DEGREE']=dc[it]
        data.loc[data.TARGET==it, 'TARGET DEGREE']=dc[it]
    return data

def inDegreeCentrality(G,data):
    idc=nx.in_degree_centrality(G)
    #print(idc)
    for it in idc:
        #print(it)
        #print(idc[it])
        data.loc[data.SOURCE==it, 'SOURCE INDEGREECENT']=idc[it]
        data.loc[data.TARGET==it, 'TARGET INDEGREECENT']=idc[it]
    return data

def outDegreeCentrality(G,data):
    odc=nx.out_degree_centrality(G)
    #print(odc)
    for it in odc:
        #print(it)
        #print(odc[it])
        data.loc[data.SOURCE==it, 'SOURCE OUTDEGREECENT']=odc[it]
        data.loc[data.TARGET==it, 'TARGET OUTDEGREECENT']=odc[it]
    return data
     
def eigenvCentrality(G,data):
    eigenvcentrality = nx.eigenvector_centrality(G)
    #print(odc)
    for it in eigenvcentrality:
        #print(it)
        #print(odc[it])
        data.loc[data.SOURCE==it, 'SOURCE EIGENVCENT']=eigenvcentrality[it]
        data.loc[data.TARGET==it, 'TARGET EIGENVCENT']=eigenvcentrality[it]
    return data

def betweennesCentrality(G,data):
    betweennescent=nx.betweenness_centrality(G)
    #print(odc)
    for it in betweennescent:
        #print(it)
        #print(odc[it])
        data.loc[data.SOURCE==it, 'SOURCE EIGENVCENT']=betweennescent[it]
        data.loc[data.TARGET==it, 'TARGET EIGENVCENT']=betweennescent[it]
    return data

def betweennesCentrality(G,data):
    betweennescent=nx.betweenness_centrality(G)
    #print(odc)
    for it in betweennescent:
        #print(it)
        #print(odc[it])
        data.loc[data.SOURCE==it, 'SOURCE EIGENVCENT']=betweennescent[it]
        data.loc[data.TARGET==it, 'TARGET EIGENVCENT']=betweennescent[it]
    return data

def clusteringCoef(G,data):
    clusteringCoef=nx.clustering(G)
    #print(odc)
    for it in clusteringCoef:
        #print(it)
        #print(odc[it])
        data.loc[data.SOURCE==it, 'SOURCE CLUSTERCOEF']=clusteringCoef[it]
        data.loc[data.TARGET==it, 'TARGET CLUSTERCOEF']=clusteringCoef[it]
    return data

def setConstraint(G,data):
    cons = nx.constraint(G)
    print(cons)
    for it in cons:
        #print(it)
        #print(odc[it])
        data.loc[data.SOURCE==it, 'SOURCE CONSTRAINT']=cons[it]
        data.loc[data.TARGET==it, 'TARGET CONSTRAINT']=cons[it]
    return data

def setEffectiveSize(G,data):
    esize = nx.effective_size(G)
    print(esize)
    for it in esize:
        #print(it)
        #print(odc[it])
        data.loc[data.SOURCE==it, 'SOURCE EFFECSIZE']=esize[it]
        data.loc[data.TARGET==it, 'TARGET EFFECSIZE']=esize[it]
    return data



G = nx.DiGraph()

rellenarGrafo(G,data)
A=nx.adjacency_matrix(G)
print(A.toarray())
print(addDegreeCentrality(G,data))
print(inDegreeCentrality(G,data))
print(outDegreeCentrality(G,data))
print('eigenvcentrality',eigenvCentrality(G,data))
print('betweennescent',betweennesCentrality(G,data))
#clicke=nx.find_cliques(G)
print('clusteringCoef',clusteringCoef(G,data))
print('cons',setConstraint(G,data))
print('esize',setEffectiveSize(G,data))
```
